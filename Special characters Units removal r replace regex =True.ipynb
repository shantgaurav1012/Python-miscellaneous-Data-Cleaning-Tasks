```python
import pandas as pd 
```

```python
df = pd.read_csv('used_cars_data.csv')
df.head()
```

```python
df.info()
```

```python
df['Power'] = df['Power'].replace({'bhp' : ''}, regex=True)
```

```python
df['Engine'] = df['Engine'].replace({'CC' : ''}, regex=True)
df['Mileage'] = df['Mileage'].replace({'kmpl': ''}, regex=True)
df.head()
```

```python
df1= pd.read_csv('Googlestore.csv')
df1.head()
```

```python
df1['Installs']=df1['Installs'].apply(lambda x : x.strip('+').replace(',', ''))
df1.head()
```

```python
df1['Size']=df1['Size'].apply(lambda x : x.strip('M'))
df1.head(2)
```

```python
df1['Android Ver'] = df1['Android Ver'].replace(r'up', 'down', regex=True)
df1.head(2)
```
